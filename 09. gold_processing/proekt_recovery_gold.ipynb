### Шаг 2.1. Анализ концентрации металлов (Ag, Pb, Au).

Создадим функцию, принимающую аргументами список с данными по концентрации на каждом этапе обработки металла, список с заголовками графиков, список с легендами и заголовок значений по оси Х. Функция строит гистограмму распределения концентрации по металлу на каждом этапе. 

```python
# создание функции для построения гистограмм распределения концентрации металла на каждом этапе обработки
def hist_met_concentr(list_met, header, legend, x_labe):
    fig, ax = plt.subplots( figsize=(12,6))
    fig = plt.figure(figsize=(16,8))
    for i in range(len(list_met)):
        df_fll.hist(list_met[i], ax=ax, grid='True', bins=50, alpha=0.4, rwidth=0.4, legend=True)
        ax.set_title(header, fontsize=16)
        ax.set_xlabel(x_labe, fontsize=14)
        ax.set_ylabel('Частота', fontsize=14)
        ax.legend(['Концентрация в исходном сырье', 'Концентрация после флотации', 'Концентрация после первичной очистки',
                  'Концентрация в готовом продукте'])
```

```python
# вызов наименований столбцов с концентрацией металлов на кажом этапе
cols_metals = [x for x in df_fll.columns if 'concentrate' in x]
cols_metals
```

```python
# создание необходимых списков для построения графиков
# список с этапами для серебра
list_ag = ['rougher.input.feed_ag', 'rougher.output.concentrate_ag', 'primary_cleaner.output.concentrate_ag',
           'final.output.concentrate_ag'] 
# список с этапами для свинца
list_pb = ['rougher.input.feed_pb', 'rougher.output.concentrate_pb', 'primary_cleaner.output.concentrate_pb',
           'final.output.concentrate_pb'] 
# список с этапами для золота
list_au = ['rougher.input.feed_au', 'rougher.output.concentrate_au', 'primary_cleaner.output.concentrate_au',
           'final.output.concentrate_au']
# список со списками по каждому металлу
list_gen = [list_ag, list_pb, list_au]
# список с наименованиями заголовков           
list_name_header = ['Изменение концентрации серебра', 'Изменение концентрации свинца', 'Изменение концентрации золота']
# список с наименованиями легенд для графиков
list_legend = ['Концентрация в исходном сырье', 'Концентрация после флотации', 'Концентрация после первичной очистки',
                  'Концентрация в готовом продукте']
```

```python
# вызов функции для построения гистограммы распределения концентрации на этапах в цикле
for j in range(3):
    hist_met_concentr(list_gen[j], list_name_header[j], list_legend, 'Концентрация')
```

 - Концентрация серебра после флотации растет примерно в полтора раза, после первичной обработки концентрация сравнивается с концентрацией в исходном сырье, концентрация в готовом продукте в среднем даже ниже, чем в исходном сырье.
 - Содержание свинца от этапа к этапу незначительно растет, а в готовом продукте в некоторых партиях концентрация даже уступает пост-первичной обработке.
 - Концентрация золота от этапа к этапу обработки существенно растет, гистограммы распределения имеют лишь незначительные пересечения с предудущими графиками, лишь в хвосте слева. Распределения имеют нормальный характер.

Распределение концентрации имеют длинные "хвосты", а также значительную частоту в значениях около нуля. Проверим эти данные на выбросы, которые могут ухудшить обучение модели и исказить результаты ее работы. Построим боксплоты разброса значений концентраций металлов.
```python
# создание функции для отрисовки боксплотов концентрации металлов 
def boxplot_met(list_met):
    fig, ax = plt.subplots(1,4, figsize=(14,4))
    for i in range(len(list_met)):
        sns.boxplot(data=df_trn, x=list_met[i], ax=ax[i])
        ax[i].set_title('Разброс данных', fontsize=12)
```

```python
# вывод на экран размеров обучающих выборок
print('Размер тренировочной выборки:', X_r.shape, X_f.shape, y_r.shape, y_f.shape)
```

Для валидации моделей будем использовать метрику SMAPE, то есть симметричный средний процент отклонения. Она позволяет увидеть результат работы модели наглядно, в относительных величинах, что удобно даже при прогнозировании величин близких к нулю. Ведь целевые признаки в нашей задаче выражены в процентах.

Создадим функцию для подсчета метрики sMape, которая принимает аргументами целевой признак и предсказания модели, а возвращает sMAPE для каждого целевого признака.

```python
# создание функции для подсчета метрики sMape
def smape(target, prediction):
    return 1/len(target) * np.sum(2*np.abs(prediction-target)/(np.abs(target)+np.abs(prediction))*100)
```

Напишем функцию для подсчета метрики итоговой sMAPE, которая принимает аргументами sMAPE для предсказаний модели эффективности обогащения чернового концентрата и sMAPE для предсказаний модели эффективности обогащения финального концентрата, а возвращает итоговый sMAPE.

```python
# создание функции для подсчета метрики итоговой sMape
def weighted_smape(smape_rougher, smape_final):
    return 0.25*smape_rougher + 0.75*smape_final
```

```python
# создание "pipeline" для трех моделей МО
pipe_lr = Pipeline([('scaler', StandardScaler()),
                    ('reg', LinearRegression())])

pipe_dr = Pipeline([('scaler', StandardScaler()),
                    ('reg', DecisionTreeRegressor(random_state=0))])

pipe_rf = Pipeline([('scaler', StandardScaler()),
                    ('reg', RandomForestRegressor(random_state=0))])
```

```python
# создание списков для перебора числовых параметров моделей МО
param_range = list(range(4, 15, 2))
param_range_1 = list(range(40, 81, 40))
# создание списков-словарей с параметрами моделей МО

param_lr = [{'reg__fit_intercept': [True, False],
             'reg__normalize': [True, False],
             'reg__copy_X': [True, False]}]

param_dr = [{'reg__max_depth': param_range,
             'reg__splitter': ['best', 'random'],
             'reg__max_features': ['sqrt']}]

param_rf = [{'reg__max_depth': param_range,
             'reg__max_features' : ['sqrt'],
             'reg__n_estimators': param_range_1}]
```

```python
# создание переменных с записью в них GridSearchCV (далее GSCV) для поиска оптимальных параметров по каждой модели МО
jobs = -1

LR = GridSearchCV(estimator=pipe_lr, param_grid=param_lr) 

DR = GridSearchCV(estimator=pipe_dr, param_grid=param_dr)

RF = GridSearchCV(estimator=pipe_rf, param_grid=param_rf)
```

### Шаг 3.2. Выбор оптимальной модели.

Напишем функцию для кросс-валидации модели МО и вывода sMAPE и итоговой sMAPE. Функция принимает модель, признаки и целевые признаки для предсказания чернового и финального концентрата, а также наименование модели. Внутри цикла поочередно выбираются индексы для валидационной и обучающей выборок, кросс-валидация происходит в 5 этапов. На каждой итерации модель обучается, делает предсказания, рассчитывается sMAPE с разными индексами данных для каждой из выборок. В заключении работы функции рассчитывается итоговый sMAPE.

```python
# создание функции для кросс-валидации модели и вывода на экран итоговой sMAPE
def cross_val_model(regressor, feature_r, feature_f, target_r, target_f, name_regressor):
    scores = []
    sample_size = int(len(feature_r) / 5)
    for i in range (0, (len(feature_r)), sample_size):
# выбор индексов для валидационной и обучающей выборок при кросс-валидации
        valid_ind = list(range(i, i+sample_size))
        train_ind = feature_r.query('index not in @valid_ind').index
# сохранение тренировочных выборок       
        features_train_r = feature_r.iloc[train_ind,:]
        features_train_f = feature_f.iloc[train_ind,:]
        target_train_r = target_r.iloc[train_ind]
        target_train_f = target_f.iloc[train_ind]
 # сохранение валидационных выборок       
        feature_valid_r = feature_r.iloc[valid_ind,:]
        feature_valid_f = feature_f.iloc[valid_ind,:]
        target_valid_r = target_r.iloc[valid_ind]
        target_valid_f = target_f.iloc[valid_ind]
# обучение модели и предсказания, а также рассчет sMAPE для чернового концентрата     
        model_r = regressor
        model_r.fit(features_train_r, target_train_r)
        predictions_r = model_r.predict(feature_valid_r)
        smape_r = smape(target_valid_r, predictions_r)
# обучение модели и предсказания, а также рассчет sMAPE для финального концентрата
        model_f = regressor
        model_f.fit(features_train_f, target_train_f)
        predictions_f = model_f.predict(feature_valid_f)
        smape_f = smape(target_valid_f, predictions_f)
# рассчет итоговой sMAPE для модели
        total_smape = weighted_smape(smape_r, smape_f)
        scores.append(total_smape)
    print('Оценки SMAPE на кросс-валидации для модели {}: {}'.format(name_regressor, scores))
    print()
    final_total_smape = np.mean(scores)
    print('Средняя оценка итоговой SMAPE для модели {}: {:.4}%'.format(name_regressor, final_total_smape))
    return final_total_smape
```

модель **LinearRegression**

```python
# вызов функции для кросс-валидации LinearRegression
cross_val_model(LR, X_r, X_f, y_r, y_f, 'LinearRegression')
```

модель **DecisionTreeRegressor**

```python
# вызов функции для кросс-валидации DecisionTreeRegressor
cross_val_model(DR, X_r, X_f, y_r, y_f, 'DecisionTreeRegressor')
```

модель **RandomForestRegressor**

```python
# вызов функции для кросс-валидации RandomForestRegressor
cross_val_model(RF, X_r, X_f, y_r, y_f, 'RandomForestRegressor')
```

Лучшая метрика итоговой SMAPE показана моделью **LinearRegression**. Далее проверим ее работу на тестовой выборке.
