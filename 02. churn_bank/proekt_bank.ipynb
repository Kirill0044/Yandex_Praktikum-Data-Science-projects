### Шаг 2.2 Проверка данных на аномалии и исправления.

**Глядя на первоначальные данные в столбце со стажем, которые должны быть представлены в днях, можно понять, что эти данные не являются корректными, так как они вещественные, а не целые числа и среди них много отрицательных значений. Но кроме этого, выделяются некоторые строки с шестизначными числами, которых не должно быть. Проверим эти данные, пересчитав дни в годы.**

```python
# усредненный пересчет в столбце о стаже данных из дней в годы и сохранение их в переменную
df_year_employed = df['days_employed']/365
df_year_employed.head(20)
```

**В данных выделяются некоторые строки в которых стаж в пересчете на годы приблизительно 932 и 1096 лет. Эти данные являются также аномальными и требуют изучения. Напишем функцию, которая принимает аргументами списки со стажем и возрастом заемщиков, а возвращает список с данными о стаже и возрасте заемщиков банка и их количество. Внутри функции условием сортировки будет стаж больше 100 лет, который в реальной жизни не может существовать.** 

```python
# создание функции, возвращающей список с аномальным стажем и возрастом клиента, а также количеством строк аномальных данных 
def count_wrong_days(days , years):
    list_wrong_days = []
    count_wrong = 0
    for index in range(len(days)):
        if days[index]/365 > 100:
            count_wrong += 1
            list_wrong_days.append([int(days[index]/365) , years[index]])
    return count_wrong , list_wrong_days 
```

```python
# формирование списков со стажем и возрастом клиентов, выведение результатов работы фунции на экран
list_days_employed = list(df['days_employed'])
list_dob_years = list(df['dob_years'])
count_wrong_days(list_days_employed , list_dob_years)
```

**Изучив часть полученного списка, можно резюмировать, что аномальные цифры стажа в подавляющем большинстве наблюдаются у заемщиков в возрасте от 50 лет. Возможно такие данные были загружены в неправильном формате или имел место сбой. К тому же данные были в дробном формате с большим числом отрицательных значений. Поэтому необходимо уточнить у разработчиков какова причина такого формата. Но так как данные о стаже не являются необходимыми для исследования, можно их не преобразовывать в дальнейшем.**

### Шаг 2.6. Категоризация дохода.

```python
# создание функции, которая принимает аргументом строку датафрейма, а возвращает категорию клиента по уровню дохода
def income_category(row):
    if row['total_income'] <30000:
        return 'E'
    elif row['total_income'] <50000:
        return 'D'
    elif row['total_income'] < 200000:
        return 'C'
    elif row['total_income'] < 1000000:
        return 'B'
    else:
        return 'A'
```

```python
# создание нового столбца total_income_category в датафрейме с помощью применения к датафрейму методом apply  функции
# income_category
df['total_income_category'] = df.apply(income_category , axis=1)
df.loc[df['total_income'] >1000000].head(10)
```

### Шаг 2.7. Категоризация целей кредита.

**Чтобы выяснить как разные цели кредита влияют на возврат его в срок, укрупним категории по целям кредита. Для этого создаем функцию, которая принимает аргументом строку датафрейма, а возвращает строку с названием категории цели кредита. Данные запишем в новый столбец.**

```python
# создание функции, возвращающей название категории цели кредита
def category_purpose(row):
    if  'жил' in row['purpose']:
        return 'операции с недвижимостью'
    elif 'недвиж' in row['purpose']:
        return 'операции с недвижимостью'
    elif 'авто' in row['purpose']:
        return 'операции с автомобилем'
    elif 'свад' in row['purpose']:
        return 'проведение свадьбы'
    elif 'образ' in row['purpose']:
        return 'получение образования'
```

```python
# создание нового столбца purpose_category в датафрейме с помощью применения к датафрейму методом apply функции category_purpose
df['purpose_category'] = df.apply(category_purpose , axis=1)
df.head(10)
```

### Шаг 2.8. Категоризация заемщика по количеству детей.

**Для того, чтобы выяснить зависимость между количеством детей и возвратом кредита в срок, будет удобнее создать категории заемщиков по количеству детей. Предварительно декомпозируем общий датафрейм для дальнейшего анализа.**

```python
# декомпозиция общего датафрейма в датафрейм для анализа данных
df_final_category = df[['children' , 'family_status_id', 'total_income_category' , 'purpose_category' , 'debt']]
df_final_category.head(10)
```

**Создадим функцию, которая принимает аргументом строку датафрейма, а возвращает строку с категорией клиента по количеству детей.**

```python
# создание функции, возвращающей название категории заемщика по количеству детей
def category_child(row):
    if row['children'] == 0:
        return 'бездетные'
    elif row['children'] >0 and row['children']<3:
        return 'с детьми'
    elif row['children'] >=3:
        return 'многодетные'
```

```python
# создание нового столбца child_category в датафрейме с помощью применения к датафрейму методом apply  функции category_child
df_final_category['child_category'] = df_final_category.apply(category_child , axis=1)
df_final_category.head(10)
```
