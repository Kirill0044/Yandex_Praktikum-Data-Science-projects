  - "M"- "Для взрослых 17+"
  - "T" - "Для подростков 13+"
  - "K-A" - "Для детей и взрослых"
  - "E10+" - "Для всех от 10 лет и старше"
  - "E" -  «Для всех»
  - "EC" - "Для детей младшего возраста"
  - "RP" - "Рейтинг ожидается".

### Шаг 3.1. Определение актуального периода и списка игровых платформ для исследования.

Изучим количество релизов в каждый год, для этого построим линейный график: ось Х - год релиза, ось Y - количество игр в год. Для этого сведем данные за каждый год методом pivot_table() и применим метод plot из библиотеки matplotlib.

```python
# постройка линейного графика с количеством игр по годам с применением pivot_table и функции count()
df.pivot_table(index='year_of_release', values='name', aggfunc='count').plot( style='o-', figsize=(10, 5), grid=True)
plt.title('Количество игр в год', fontsize=16)
plt.xlabel('Год', fontsize=14)
plt.ylabel('Количество игр', fontsize=14)
```

Выпуски игр до 1993 года можно сказать, что не увеличивалось. Затем, видимо с появлением новых консолей и расширением рынков продажи, количество увеличивалось постепенно. Резкий рост релизов начался с 2000 года, после 2002 даже наблюдался небольшой спад, но с 2004 года начался резкий подъем в количестве выпускаемых игр. Скорее всего немаловажную роль здесь сыграло электронное распространение контента игр и продажа ключей игр через сеть("Steam" открыла продажи в конце 2003 года). После 2008 года пошел спад в релизах, что может быть связано с увеличением продаж игр для мобильных устройств. После 2013 года продолжается незначительный спад в количестве релизов. **Исходя из этих данных возьмем для исследования период, когда шел последний резкий рост в количестве выпускаемых игр, который затем продолжился еще большим спадом в активности игроделов. Определим первоначальный период для исследования с 2004 года и запишем его в переменную.**

```python
# запись первоначального  периода для исследования в переменную
df_actual_per_first = df.query('year_of_release > 2003')
```

Определим наиболее прибыльные платформы для исследования, для этого сведем данные по продажам для каждой платформы методом pivot_table() и построим столбчатую диаграмму.

```python
# построение столбчатой диаграммы с применением pivot_table() и plot()
df_actual_per_first.pivot_table(index='platform', values='total_sales', aggfunc='sum').sort_values(by='total_sales', 
                                 ascending=False).plot(kind='bar', grid=True, figsize=(12,6))
plt.title('Объем продаж по игровым платформам', fontsize=16)
plt.xlabel('Платформа', fontsize=14)
plt.ylabel('Объем продаж', fontsize=14)
plt.xticks(fontsize=12)
```

Как видим, выделяется первая пятерка по продажам за актуальный период: XBOX_360, PS3, Wii, Ds, PS2. Среди лидеров по продажам есть не самые "свежие" платформы, поэтому для исследования возьмем еще несколько платформ, включая новую Xone.

```python
# группировка данных по глобальным продажам на каждой платформе методом pivot_table()
big_sales_platf =  df_actual_per_first.pivot_table(index='platform', values='total_sales',
                                  aggfunc='sum').sort_values(by='total_sales', ascending=False).head(10)
```

```python
# запись наименований платформ в список
list_big_platform = list(big_sales_platf.index)
list_big_platform
```

Построим линейные графики по глобальным продажам на каждой ведущей платформе за актуальный период. Для этого построим все графики на одной оси координат, применяя цикл for и pivot_table, сведя данные по продажам по каждой платформе.

```python
# построение линейных графиков по глобальным продажам в каждый год на каждой платформе, применяя цикл и pivot_table() и plot()
fig, ax = plt.subplots( figsize=(16,8))
fig = plt.figure(figsize=(16,8))
for i in list_big_platform:
    ax = df_actual_per_first.query('platform == @i').pivot_table(index='year_of_release', values='total_sales', 
                                                    columns='platform', aggfunc='sum').plot( ax=ax, style='o-', grid='True')
    ax.set_title('Объем продаж по платформам', fontsize=16)
    ax.set_xlabel('Год продажи игр', fontsize=14)
    ax.set_ylabel('Объем продаж, млн.', fontsize=14)
```

Из графиков видно, что каждая из платформ имеет свой срок существования: с начала и роста продаж до полного прекращения выпуска и продажи игр для нее. Оценки критиков влияют на продажи ощутимо сильнее, хотя их влияние все еще не является определяющим фактором для финансового успеха. Построим диаграмму рассеяния.

По этой диаграмме уже ясно, что оценки выше 60 баллов из 100 влекут только увеличение продаж, особенно это заметно на 80 баллах и выше, где уже значимая часть игр продается в более чем 2 млн. экземплярах. Построим диаграммы рассеяния и посчитаем коэффициенты корреляции для других платформ. Для этого в цикле переберем все актуальные платформы и методом scatter построим диаграммы зависимости глобальных продаж от оценки критиков и пользовательских оценок. При этом составим условия внутри цикла по разным числам коэффициэнтов, чтобы вывести на экран степень зависимости величин, не будем принимать во внимание отрицательные коэффициэнты(маочень маловероятна существенная обратная зависимость показателей).

```python
# построение диаграмм рассеяния по влиянию оценок пользователей и критиков на глобальные продажи на всех актуальных платформах,
# а также рассчет соответствующих коэффициэнтов корреляции и вывод на экран степеней взаимозависимости показателей
for i in list_actual_platf:
    platf = df_actual_per[df_actual_per['platform'] == i]
    fig, ax = plt.subplots(1 ,2, figsize=(15,5))
    ax[0].scatter(data=platf, x='critic_score', y='total_sales',  color='#9467bd', alpha=0.3)
    ax[1].scatter(data=platf, x='user_score', y='total_sales',  color='#1f77b4', alpha=0.3)
    fig.suptitle(i, fontsize=14)
    ax[0].set_xlabel('Оценка игр критиками', fontsize=13)
    ax[1].set_xlabel('Оценка игр пользователями', fontsize=13)
    ax[0].set_ylabel('Объем продаж, млн.', fontsize=13)
    ax[1].set_ylabel('Объем продаж, млн.', fontsize=13)  
    plt.show()
    # запись в переменные коэффициентов корреляции по оценкам пользователей и критиков 
    cor_crit = platf['critic_score'].corr(platf['total_sales'])
    cor_user = platf['user_score'].corr(platf['total_sales'])
    # запись внутри цикла условий и выводов по влиянию отзывов на глобальные продажи
    if  cor_crit <= 0.3:
        print(f'Коэф. корел-и: {cor_crit}. Влияние оценки критиков на объем продаж игр на платформе {i}  слабое.')
    if  0.3 <= cor_crit <= 0.5:
        print(f'Коэф. корел-и: {cor_crit}. Влияние оценки критиков на объем продаж игр на платформе {i} заметное.')
    if 0.5 <= cor_crit <= 0.7:
        print(f'Коэф. корел-и: {cor_crit}. Влияние оценки критиков на объем продаж игр на платформе {i} среднее.')
    if 0.7 < cor_crit:
        print(f'Коэф. корел-и: {cor_crit}. Влияние оценки критиков на объем продаж игр на платформе {i} сильное.') 
    if  cor_user <= 0.3:
        print(f'Коэф. корел-и: {cor_user}. Влияние оценки пользователей на объем продаж игр на платформе {i}  слабое.')
    if  0.3 <= cor_user <= 0.5:
        print(f'Коэф. корел-и: {cor_user}. Влияние оценки пользователей на объем продаж игр на платформе {i} заметное.')
    if 0.5 <= cor_user <= 0.7:
        print(f'Коэф. корел-и: {cor_user}. Влияние оценки кпользователей на объем продаж игр на платформе {i} среднее.')
    if 0.7 < cor_user:
        print(f'Коэф. корел-и: {cor_user}. Влияние оценки пользователей на объем продаж игр на платформе {i} сильное.')
   
```

**Из данных графиков видно, что связь между оценками пользователей и объемом продаж видеоигр на всех платформах практически отсутствует, а на консолях нового поколения она сведена к нулю(разница возможно сказывается с еще не накопленными данными в достаточном объеме за период существования новых консолей). На платформе DS  оценки критиков не сказываются на увеличении продаж, чего не скажешь о консолях Playstation и Xbox и Wii, как нового, так и старого поколения, здесь несколько заметна направленность на увеличение продаж при хороших оценках, это еще более заметно на отметках от 80 и выше.**

### Шаг 3.4. Анализ распределения игр по жанрам и их прибыльности.

Сведем методом pivot_table() в отдельные таблицы данные по количеству игр в отдельных жанрах и глобальных продаж игр в этих жанрах. Также затем мы сможем посчитать и средние продажи 1 игры по каждому жанру.

С большим, более чем двукратным преимуществом лидирует жанр Action, далее почти поровну sports и misc, а вот меньше всего выпустили стратегий и игр-головоломок.

По продажам уже не все так однозначно, первенство все также у жанра Action, но отрыв от Sports уже всего около 30 %, а следом хорошо продаются игры жанра shooter, а стратегии и головоломки также на последних местах(только поменялись местами). Далее создадим таблицу, в которой методом join() объединим по названиям жанров(которые являются индексами) две предыдущие таблицы. Также рассчитаем средний показатель объема глобальных продаж с 1 выпущенной игры в конкретном жанре и запишем эти данные в новый столбец.

```python
# объединение таблиц с количеством игр и объемами продаж по жанрам в одну и добавление столбца с данными о среднем показателе
# объема глобальных продаж с 1 выпущенной игры в конкретном жанре 
df_genr_profit = df_genr_count.join(df_genr_sum)
df_genr_profit['profit'] = df_genr_profit['total_sales'] / df_genr_profit['name']
df_genr_profit
```

Построим столбчатую диаграмму по средним показателям объема глобальных продаж с 1 выпущенной игры по каждому жанру.

## Шаг 4. Составление портрета пользователя каждого региона.

Для более точного представления об средней успешной игре разберем территориальные предпочтения покупателей в основных регионах продаж. Выясним какие наиболее популярные платформы, игровые жанры в регионе. Также посмотрим какой возраст у пользователя преобладает в каждом регионе. Чтобы учесть все последние тенденции, проанализируем отдельно период 2016 года. Для реализации этого подойдет построение столбчатых диаграмм для каждого случая. Воспользуемся созданием функции, которая принимает списки с названием региона и наименованиями этих регионов, внутри которой в цикле создаются сводные таблицы с индексами жанров, платформ и рейтингов ESRB, а данные продаж по каждому региону суммируются. Далее идет построение столбчатых диаграмм.

```python
# создание фунции, внутри которой в теле цикла по каждому показателю сводится таблица с данными глобальных продаж за актуальный 
# период и строится столбчатая диаграмма 
def region_bar(list_reg, header):
    for i in range(len(list_reg)): 
        fig, ax = plt.subplots(1, 3, figsize=(14, 4))
        ax[0] = df_actual_per.pivot_table(index='genre', values=list_reg[i], aggfunc='sum').sort_values(by=list_reg[i], 
                                        ascending=False).head(5).plot(kind='bar', ax=ax[0], color=(0.1, 0.8, 0.6, 0.9))
        ax[1] = df_actual_per.pivot_table(index='platform', values=list_reg[i], aggfunc='sum').sort_values(by=list_reg[i],
                                        ascending=False).head(5).plot(kind='bar', ax=ax[1], color=(0.2, 0.5, 0.8, 0.7))
        ax[2] = df_actual_per.pivot_table(index='rating', values=list_reg[i], aggfunc='sum').sort_values(by=list_reg[i],
                                        ascending=False).head(5).plot(kind='bar', ax=ax[2], color=(0.6, 0.5, 0.8, 0.7))
        fig.suptitle(f'Продажи по жанрам, платформам и рейтингу ESRB в регионе: {header[i]}', fontsize=18)
        ax[0].set_xlabel('Жанр игры', fontsize=15)
        ax[1].set_xlabel('Игровая платформа', fontsize=15)
        ax[2].set_xlabel('Рейтинг ESRB', fontsize=15)
        ax[0].set_ylabel('Объемы продаж, млн.', fontsize=14)
        ax[1].set_ylabel('Объемы продаж, млн.', fontsize=14)
        ax[2].set_ylabel('Объемы продаж, млн.', fontsize=14)
        ax[0].tick_params(axis='x', labelrotation=45, labelsize=14)
        ax[1].tick_params(axis='x', labelrotation=45, labelsize=14)
        ax[2].tick_params(axis='x', labelrotation=45, labelsize=14)
    print('\n')     
```

```python
# создание списков с наименованиями регионов и заголовками, применение функции по построению диаграмм
list_region = ['na_sales', 'eu_sales', 'jp_sales']
headers = ['Северная Америка', 'Европа', 'Япония']
region_bar(list_region, headers)
```

## Шаг 5. Проверка гипотез.

### Шаг 5.1. Проверка гипотезы о равенстве средних пользовательских рейтингов платформ XBOX_one и PC

Нулевую гипотезу обычно формулируют исходя из равенства значений сравниваемых совокупных выборок. Поэтому сформулируем нулевую гипотезу следующим образом:

Проверка первой гипотезы. **H0: средние пользовательские рейтинги на платформах XBOX_one и PC равны.** Альтернативную гипотезу сформулируем как двухстороннюю: **Н1: средние пользовательские рейтинги на платформах XBOX_one и PC различаются.** Для того, чтобы отвергнуть или подтвердить гипотезу по 2-м выборкам будем использовать метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, то есть двухсторонний ttest методом scipy.stats.ttest_ind(array1, array2, equal_var). Рассчитаем предварительно среднее значение по выборкам, дисперсию и стандартное отклонение. Если дисперсии распределения у игр двух платформ отличаются, то укажем параметр equal_var=False. Предварительно запишем массивы выборок по платформам в отдельные переменные. Определим параметр значимости alpha = 0.01.

```python
# сохранение данных по оценкам пользователей на каждой платформе в переменные
xbox_one_user_score = df_actual_per[df_actual_per['platform'] == 'xone']['user_score']
pc_user_score = df_actual_per[df_actual_per['platform'] == 'pc']['user_score']
```

```python
# удаление пропусков в данных по XBOX_one
xbox_one_user_score = xbox_one_user_score.dropna()
xbox_one_user_score.head(10)
```

```python
# удаление пропусков в данных по PC
pc_user_score = pc_user_score.dropna()
pc_user_score.head(10)
```

```python
# рассчет среднего пользовательских оценок по обеим платформам
print('Средняя пользовательская оценка на платформе XBOX_ONE: {:.1f} балла.'.format(xbox_one_user_score.mean()))
print('Средняя пользовательская оценка на платформе PC: {:.1f} балла.'.format(pc_user_score.mean()))
```

```python
# рассчет дисперсии распределения пользовательских оценок по обеим платформам
variance_xbox_one_user_score = np.var(xbox_one_user_score)
print('Дисперсия по пользовательским оценкам на платформе XBOX_ONE: {:.1f}'.format(variance_xbox_one_user_score))
variance_pc_user_score = np.var(pc_user_score)
print('Дисперсия по пользовательским оценкам на платформе PC: {:.1f}'.format(variance_pc_user_score))
```

```python
# рассчет стандартного отклонения пользовательских оценок по обеим платформам
std_dev_xbox_one_user_score = np.std(xbox_one_user_score)
print('Стандартное отклонение по пользовательским оценкам на платформе XBOX_ONE: {:.1f}'.format(std_dev_xbox_one_user_score))
std_dev_pc_user_score = np.std(pc_user_score)
print('Стандартное отклонение по пользовательским оценкам на платформе PC: {:.1f}'.format(std_dev_pc_user_score))
```

Дисперсии по Платформам XBOX_ONE и PC различаются, укажем для ttest_ind параметр equal_var=False.

```python
# проверка двусторонней гипотезы методом st.ttest_ind
alpha = 0.01,
results = st.ttest_ind(xbox_one_user_score, pc_user_score, equal_var=False)

print('p-значение:', '{:.8f}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средних пользовательских оценок на платформах XBOX_ONE и PC.')
else:
    print('Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских оценок на платформах XBOX_ONE и PC.')
```

Вероятность встретить средние одинаковые оценки у пользователей платформ XBOX_ONE и PC достаточно велика и составляет около 16 % , поэтому принимаем нулевую гипотезу о равенстве средних оценок пользователей этих платформ.

Для того, чтобы отвергнуть или подтвердить гипотезу по 2-м выборкам будем использовать метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, а именно двухсторонний ttest методом scipy.stats.ttest_ind(array1, array2, equal_var). Рассчитаем предварительно среднее значение по выборкам, дисперсию и стандартное отклонение. Если дисперсии распределения у игр двух жанров отличаются, то укажем параметр equal_var=False. Предварительно запишем массивы выборок по платформам в отдельные переменные. Определим параметр значимости alpha = 0.01.

```python
# сохранение данных по оценкам пользователей на каждом тарифе в переменные
action_user_score = df_actual_per[df_actual_per['genre'] == 'action']['user_score']
sports_user_score = df_actual_per[df_actual_per['genre'] == 'sports']['user_score']
```

```python
# рассчет стандартного отклонения пользовательских оценок по обоим жанрам
std_dev_action_user_score = np.std(action_user_score)
print('Стандартное отклонение по пользовательским оценкам в играх жанра Action: {:.1f}'.format(std_dev_action_user_score))
std_dev_sports_user_score = np.std(sports_user_score)
print('Стандартное отклонение по пользовательским оценкам в играх жанра Sports: {:.1f}'.format(std_dev_sports_user_score))
```

```python
# проверка двусторонней гипотезы методом st.ttest_ind
alpha = 0.01,
results = st.ttest_ind(action_user_score, sports_user_score, equal_var=False)

print('p-значение:', '{:.20f}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средних пользовательских оценок в играх жанра Action и Sports.')
else:
    print('Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских оценок в играх жанра Action и Sports.')
```

Вероятность встретить средние одинаковые оценки у пользователей на играх жанров Action и Sports критически мала, поэтому отвергаем нулевую гипотезу о равенстве пользовательских оценок и принимаем альтернативную о их различии.
